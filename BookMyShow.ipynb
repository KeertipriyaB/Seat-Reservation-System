```python
class theater:
    def __init__(self):
        self.t=[]
        for i in range(5):
            a = []
            for j in range(5):
                a.append("*")
            self.t.append(a)
    def seat_matrix(self):
        print("   ", end="")
        for i in range(5):
            print(f" {i} ", end="")
        print()
        for i in range(5):
            print(f"{i} ", end="")
            for j in range(5):
                print(" ",self.t[i][j], end="")
            print()
    def book_ticket(self):
        row,column= int(input("Enter the row no")),int(input("Enter the column no"))
        price=0
        if self.t[row][column]=="*":
            self.t[row][column]="#"
            print("Ticket is booked succesfully")
        else:
            print("Seat is already taken")
        if row<2:
            price=220
        else:
            price=200
        return price
    def cancel_ticket(self):
        row,column= int(input("Enter the row no")),int(input("Enter the columns no"))
        price=0
        if self.t[row][column]=="#":
            self.t[row][column]="*"
            print("Seat has been cancelled succesfully")
        else:
            print("Seat is already empty")
        if row<2:
            price=220
        else:
            price=200
        return price

a=theater()
n=int(input("No of tickets"))
count = 0
cost=0
while True:
        choice=int(input("1.Display available seats 2.Book ticket 3.Cancel 4.Check out "))
        if choice==1:
            print("Price of Gold Seat 0&1 = Rs220")
            print("Price of Normal Seat 2-4 = Rs200")
            a.seat_matrix()
        elif choice==2:
            if count<n:
                cost+=a.book_ticket()
                count+=1
            else:
                print("Error")
        elif choice==3:
            cost-=a.cancel_ticket()
            count-=1
        elif choice==4:
            print("Rs",cost)
            print("Thank you for booking with us")
            exit()
            break
        else:
            print("Invalid entry, Please try again")
            exit()
            break
```
